```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 5)

# Step 1: Create the output/tables folder if it doesn't exis
os.makedirs('output/tables', exist_ok=True)

# Load cleaned data
df_main = pd.read_csv('../data/processed/cleaned_googleplaystore.csv')
df_main.head()

# Group and calculate average ratings
average_rating = df_main.groupby('Category')['Rating'].mean().reset_index()

# Save the table
average_rating.to_csv('output/tables/average_rating_by_category.csv', index=False)

# Optional: print preview
average_rating.head()
```
